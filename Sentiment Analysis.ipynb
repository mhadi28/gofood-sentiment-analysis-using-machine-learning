```python
import pandas as pd
import emoji
import numpy as np
import contractions
import re
import string
```

# Cleaning

```python
def remove(tweet):
    #remove tags
    t1 = re.sub('RT\s', '', tweet)
    #remove @username
    t2 = re.sub('\B@\w+', "", t1)
    #remove url\
    t3 = re.sub(r'https?:\/\/.*[\r\n]*', '', t2)reove
    #remove emoji 
    t4 = emoji.replace_emoji(t3, replace='')
    #remove word from hashtags
    t5 = re.sub('#+', '', t4)
    #conver into lowercase
    t6 = t5.lower()
    #remove looping word likes banyaaakkk
    t7 = re.sub(r'(.)\1+', r'\1\1', t6)
    #remove punctuation
    t8 = re.sub(r'[\?\/\!]+(?=[\?.\!])', '', t7)
    #remove number and special characters
    t9 = re.sub(r'[^a-zA-Z]', ' ', t8)
    # sentence reconstruction
    t10 = contractions.fix(t9)
    return t9
```

```python
df = pd.read_csv('gofood.csv', sep=',', encoding='utf-8')
df
```

```python
for i, r in df.iterrows():
    y = remove(r['tweet'])
    df.loc[i, 'tweet'] = y
    
df.head(50)
```

# Tokenize

```python
from nltk.tokenize import word_tokenize

def word(tweet):
    return word_tokenize(tweet)

df['token'] = df['tweet'].apply(word)
```

```python
df.head(5)
```

# Stopword removal

```python
from nltk.corpus import stopwords

list_stopwords = stopwords.words('indonesian')
list_stopwords.extend(['amp', 'nb', 'k', 'n', '   ', 'me', 'ih', 'pengin',
                      'di', 'ke', 'wkndjanaksn', 'g', 'ga', 'gk', 'usa', 'h', 
                       'jd', 'tp','yg', 'gw', 'aj', 's', 'rb', 'rp', 't','mv', 'link', 'nb',
                      'wtb', 'wts', 'heheheh', 'pls', 'plss', 'x', 'sj', 'ok', 'd'])
list_stopwords = set(list_stopwords)

def stopword_removal(words):
    return [word for word in words if word not in list_stopwords]

df['re_stop'] = df['token'].apply(stopword_removal)
print(df['re_stop'])
```

# Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemmed(term):
    return stemmer.stem(term)

term_dict = {}

for document in df['re_stop']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")


# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['stem'] = df['re_stop'].swifter.apply(get_stemmed_term)
print(df['stem'])
```

```python
#join sentence

for i in range(len(df['stem'])):
    df.stem[i] = ' '.join(df.stem[i])
    
df['join'] = df.stem
df.head(10)
```

```python
from textblob import TextBlob

def getPolarity(tweet):
    return TextBlob(tweet).sentiment.polarity

def analyze(score):
    if score < 0:
        return '0'
    elif score == 0:
        return '1'
    else:
        return '2'
```

```python
final_data = pd.DataFrame(df[['datetime', 'username', 'join']])
final_data.head()
```

```python
final_data['polarity'] = final_data['join'].apply(getPolarity)
final_data['score'] = final_data['polarity'].apply(analyze)
final_data.head(10)
```

```python
final_data.score.value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=10000)
vectors = vectorizer.fit_transform(final_data['join'])
words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
words_df.head()
```

```python
X = words_df
y = final_data.score
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
```

```python
%%time
# Create and train a logistic regression
logreg = LogisticRegression(C=100, solver='lbfgs', max_iter=1000)
logreg.fit(X, y)
```

```python
%%time
# Create and train a random forest classifier
forest = RandomForestClassifier(n_estimators=50)
forest.fit(X, y)
```

```python
%%time
# Create and train a linear support vector classifier (LinearSVC)
svc = LinearSVC()
svc.fit(X, y)
```

```python
%%time
# Create and train a multinomial naive bayes classifier (MultinomialNB)
bayes = MultinomialNB()
bayes.fit(X, y)
```

```python
# Put it through the vectoriser

# transform, not fit_transform, because we already learned all our words
unknown_vectors = vectorizer.transform(final_data['join'])
unknown_words_df = pd.DataFrame(unknown_vectors.toarray(), columns=vectorizer.get_feature_names())
unknown_words_df.head()
```

```python
# Predict using all our models. 

# Logistic Regression predictions + probabilities
final_data['pred_logreg'] = logreg.predict(unknown_words_df)
final_data['pred_logreg_proba'] = logreg.predict_proba(unknown_words_df)[:,1]

# Random forest predictions + probabilities
final_data['pred_forest'] = forest.predict(unknown_words_df)
final_data['pred_forest_proba'] = forest.predict_proba(unknown_words_df)[:,1]

# SVC predictions
final_data['pred_svc'] = svc.predict(unknown_words_df)

# Bayes predictions + probabilities
final_data['pred_bayes'] = bayes.predict(unknown_words_df)
final_data['pred_bayes_proba'] = bayes.predict_proba(unknown_words_df)[:,1]
```
